```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
%matplotlib inline

```

# Inferences from the file : "calendar.csv"

```python
cal_df = pd.read_csv('calendar.csv')
cal_df
```

# Question1: How was the availability of properties in Airbnb differ for each month in a single year ?

```python
# Remove rows with missing 'price' values from the 'cal_df' 
# and create a new DataFrame 'cal_mod_df' without those rows.
# Then, create a copy of 'cal_mod_df' called 'cal_mod_df_copy'.
cal_mod_df = cal_df.dropna(subset=['price'],axis=0)
cal_mod_df_copy = cal_mod_df.copy()
cal_mod_df_copy.loc[:, 'price'] = cal_mod_df_copy['price'].str.replace('$', '').str.replace(',', '').astype(float)
cal_mod_df_copy
```

```python
# Convert the 'Date' column to datetime
cal_mod_df_copy['date'] = pd.to_datetime(cal_mod_df_copy['date'])

# Extract the month from the date and create a new column
cal_mod_df_copy['Month'] = cal_mod_df_copy['date'].dt.month
cal_mod_df_copy = cal_mod_df_copy.drop('date',axis=1)
cal_mod_df_copy
```

```python
# Get unique integer values
unique_values = cal_mod_df_copy['listing_id'].unique()

# Sort the unique values
unique_values.sort()

# Create a new column with labels like 'P1', 'P2', 'P3', etc.
cal_mod_df_copy['Property'] = 'P' + (cal_mod_df_copy['listing_id'].rank(method='dense').astype(int)).astype(str)

cal_mod_df_copy = cal_mod_df_copy.drop('listing_id',axis=1)
cal_mod_df_copy
```

```python
# Grouping by month and counting unique properties
monthly_property_count = cal_mod_df_copy.groupby('Month')['Property'].nunique()

# Plotting
plt.figure(figsize=(12, 6))
sns.barplot(x=monthly_property_count.index, y=monthly_property_count.values)
plt.title('Number of Unique Properties Available Each Month')
plt.xlabel('Month')
plt.ylabel('Number of Properties')
plt.xticks(range(0, 12), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])

plt.show()
```

# Inference 1: 

    There are more properties (Listings) available at the beginning of the year, that is the January month. But, as the      months goes on, at the middle of the year there is sliding trend which can seen clearly through the months, May, June,  July.

    Yet, the trend changes from August onwards with a raising availability of properties till the end of the year.

# Question 2: If the availability is turned out to be an occupancy what would be the revenue for each month in that year?

```python
# Function to sum unique prices for each group
def sum_unique_prices(group):
    return group['price'].drop_duplicates().sum()

# Grouping by month and summing unique prices
monthly_price_sum = cal_mod_df_copy.groupby('Month').apply(sum_unique_prices)

# Plotting
plt.figure(figsize=(12, 6))
sns.barplot(x=monthly_price_sum.index, y=monthly_price_sum.values)
plt.title('Sum of Unique Prices for Each Month')
plt.xlabel('Month')
plt.ylabel('Sum of Unique Prices')
plt.xticks(range(0, 12), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])

plt.show()
```

# Inference 2:

    If the all the available places are occupied for an instance, the monthly revenue from this aspect goes sky high in June than the other months because, the costly properties are available more in June.

    In January, even though there is a lot of availability in the Airbnb properties, the revenue is low as against the       previous scenario. It has low rent listings more. 

# Question 3: How is the rent price per day vary for each properties?

```python
# Assigning a unique number to each property
cal_mod_df_copy['PropertyID'] = cal_mod_df_copy['Property'].astype('category').cat.codes

# Now plot
plt.figure(figsize=(15, 10))
plt.scatter(cal_mod_df_copy['price'], cal_mod_df_copy['PropertyID'], alpha=0.5)
plt.xlabel('Price')
plt.ylabel('Property ID')
plt.title('Price Range for Each Property')

# Adjust y-axis to show intervals
plt.yticks(range(0, len(cal_mod_df_copy['Property'].unique()), 500), range(0, len(cal_mod_df_copy['Property'].unique()), 500))

plt.show()
```

# Inference 3: 

    The scatter plot proves that majority of the properties are with rents less than $500.  

    With more confidence, we can set the limit below $250.

    This establishes that, Airbnb always promotes more affordable homes than luxury flats and apartments.

# Question 4: How the properties and rent price per day relate if people like to afford up to $100?

```python
# Filter the data to include only prices up to 200 dollars
filtered_data = cal_mod_df_copy[cal_mod_df_copy['price'] <= 100]

# Assigning a unique number to each property in the filtered dataset
filtered_data.loc[:,'PropertyID'] = filtered_data['Property'].astype('category').cat.codes

# Now plot the filtered data
plt.figure(figsize=(15, 10))
plt.scatter(filtered_data['price'], filtered_data['PropertyID'], alpha=0.5)
plt.xlabel('Price')
plt.ylabel('Property ID')
plt.title('Price Range (up to $100) for Each Property')

# Adjust y-axis to show intervals
max_id = filtered_data['PropertyID'].max()
interval = max(1, max_id // 10)  # Adjust the interval as needed
plt.yticks(range(0, max_id + 1, interval), range(0, max_id + 1, interval))

plt.show()
```

# Inference 4:

    If we take lower priced properties below $100, there are less properties below $50.

    There is an even distribution of properties againt the price between $60 and $100.

    This concretes the fact that Airbnb has the most of its properties between the price range $60 and $250.
