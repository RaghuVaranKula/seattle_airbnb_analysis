# Question 6: Which are the top 15 Listings on Airbnb with most reviews?

# Dataset: Seattle Airbnb Open Data 
link: https://www.kaggle.com/datasets/airbnb/seattle/data 

Context:
Since 2008, guests and hosts have used Airbnb to travel in a more unique, personalized way. As part of the Airbnb Inside initiative, this dataset describes the listing activity of homestays in Seattle, WA.


Content:
The following Airbnb activity is included in this Seattle dataset:

"calendar.csv" - Calendar, including listing id and the price and availability for that day     
"listings.csv" -   Listings, including full descriptions and average review score      
"reviews.csv" - Reviews, including unique id for each reviewer and detailed comments

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
%matplotlib inline

```

# Inferences from the file : "review.csv"

```python
rev_df = pd.read_csv('reviews.csv')
rev_df
```

```python
# Grouping the data by listing_id and counting unique reviewer IDs for each listing
unique_reviewer_counts = rev_df.groupby('listing_id')['reviewer_id'].nunique()

# Getting the top 15 listing IDs with the most unique reviewer counts
top_15_listings = unique_reviewer_counts.nlargest(15)
print(top_15_listings)
# Plotting
plt.figure(figsize=(12, 6))
top_15_listings.plot(kind='bar')
plt.xlabel('Listing ID')
plt.ylabel('Unique Reviewer Count')
plt.title('Top 15 Listings with the Most Unique Reviewers')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
